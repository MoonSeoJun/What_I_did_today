```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0,10,1)

# write praph
plt.plot(t,(-t), 'bd',markersize=10, linestyle='-.', label='-n')
plt.plot(t, 2 * t, "rp", markersize=13, linestyle='--', label='2n')
plt.plot(t, t**2, "g*", markersize=16, linestyle=':', label='n**2')

# write grid
plt.grid(True, axis='x', color='red', alpha=0.3)

# write ticks
plt.xticks([i for i in range(0,10)])
plt.tick_params(direction='out', length=6, width=2)

# write label and title
plt.xlabel("xxx label", color='pink', labelpad=15, size=10, fontdict={'family' : 'FONTNAME'})
plt.ylabel("yyy label", color='gray', labelpad=15, size=15, fontdict={'family' : 'fantasy'})
plt.title("plt_quiz", pad=30, loc='center')

# write legend and show
plt.legend()
plt.show()
```
